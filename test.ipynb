## Year vs Number of Movies Line Chart

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('imdb_top_1000.csv')

# Clean and drop missing values 
df["Released_Year"] = pd.to_numeric(df["Released_Year"], errors="coerce") # turn to numeric
new_df = df.dropna(subset=["Released_Year"]) # Drop missing values from dataframe

grouped = new_df.groupby("Released_Year").size() # Aggregate
summary_df = grouped.reset_index()
summary_df.columns = ["Released_Year", "Movie_Count"]

print(summary_df)

# Plot
fig = px.line(summary_df, x="Released_Year", y="Movie_Count", title="# of Movies Released Yearly")
fig.show()
```
